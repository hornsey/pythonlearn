```python
import numpy as np
import matplotlib.pyplot as plt

GDP = [12304.3,13080.3,9421.3,9102.54]

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

plt.barh(['Beijing','Shanghai','Tianjin','Chongqing'], GDP)
plt.xlabel('GDP')
plt.title('四个直辖市GDP大比拼')
#plt.xticks(range(4), ['Beijing','Shanghai','Tianjin','Chongqing'])
plt.xlim([5000,15000])

for x,y in enumerate(GDP):
    #print(x,y)
    plt.text(y+500,x, '%s' % round(y,1), ha = 'center')

#plt.close()   
```
