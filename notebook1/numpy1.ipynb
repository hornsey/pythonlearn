## 柱状图、水平柱状图

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(5)
y=np.random.randint(-5,5,5)
fig,axes = plt.subplots(2)
fig1=axes[0].bar(x,y)
fig2=axes[1].barh(x,y)
axes[0].axhline(0,color='gray')
```

```python
x=np.random.randn(100).cumsum()
y=np.linspace(0,10,100)
y1=np.linspace(0,10,100)+2

fig,ax = plt.subplots(2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,10)
ax[0].fill_between(y,x)
ax[1].fill_between(x,y,y1)
```
