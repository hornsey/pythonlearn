# 条形图外观

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
green_data = [2,3,4]
blue_data = [3,2,1]
red_data = [2,3,4]
labels = ['group1','group2','group3']

pos = list(range(len(green_data)))
width = 0.2
fig,ax = plt.subplots(figsize=(8,6))

plt.bar(pos,green_data,width,alpha=0.5,color='g',label=labels[0])
plt.bar([p+width for p in pos],blue_data,width,alpha=0.5,color='b',label=labels[1])
plt.bar([p+width*2 for p in pos],red_data,width,alpha=0.5,color='r',label=labels[2])
```

# 填充样式

```python
patterns = ('-','+','x','\\','*','o','O','.') 
flg = plt.gca()

mean_value = range(1,len(patterns)+1)
x_pos = list(range(len(mean_value)))

bars = plt.bar(x_pos,mean_value)

for bar,pattern in zip(bars,patterns):
    bar.set_hatch(pattern)
```

# 盒图

```python
hetu_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(hetu_data,notch=False,sym='s',patch_artist=True)

plt.xticks([y+1 for y in range(len(hetu_data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plt')

colors = ['pink','lightblue','lightgreen']
for pathch,color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
```

# 小提琴图violinplot

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
vio_data = [np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(vio_data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')

axes[1].boxplot(vio_data)
axes[1].set_title('box_plot')

labels = ['x1','x2','x3','x4']


for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(vio_data))])

plt.setp(axes,xticks=[y+1 for y in range(len(vio_data))],xticklabels=labels)
```
