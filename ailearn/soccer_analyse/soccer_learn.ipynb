## 足球运动员数据分析学习

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
player = pd.read_csv("FullData.csv")
pd.set_option("max_columns", 100)
```

```python
player = player[player["Club_Kit"].notnull()]
```

```python
np.__version__
```

```python
player.head()
```

```python
player.info()
```

```python
player.describe()
```

```python
player.duplicated().any()
player.drop_duplicates(inplace=True)
```

```python
player["Height"].head()
```

```python
player["Height"] = player["Height"].map(lambda x:int(x.replace("cm", "")))
```

```python
player["Weight"].head()
```

```python
player["Weight"] = player["Weight"].str.replace("kg","").astype(np.int)
```

```python
player["Weight"].head()
```

```python
player.info()
```

运动员身高、体重和评分信息分布

```python
player[["Height", "Weight", "Rating"]].plot(kind="kde")
```

```python
player["Preffered_Foot"].value_counts().plot(kind="bar")
```

top10评分的俱乐部/国家

```python
s = player.groupby("Club")
```

```python
s.head()
```

```python
s = player.groupby("Club")["Rating"].agg(["count", "sum", "mean"])
```

```python
s.info()
```

```python
s = s[s["count"] > 20]
```

```python
s.info()
```

```python
s.sort_values("mean", ascending=False).head(10)
```

```python
s2 = player.groupby("Nationality")["Rating"].agg(["count", "sum", "mean"])
```

```python
s2.info()
```

```python
s2 = s2[s2["count"] > 20]
s2.sort_values("mean", ascending=False).head()
```

```python
s2.info()
```
